```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Menampilkan Data Klasifikasi Segmentasi Kostumer

```python
df = pd.read_csv("Test.csv")

df
```

## Visualisasi Jumlah Data Pada Attribut Segmentation Menggunakan Grafik Batang
---
Dari hasil visualisasi data berikut ini, dapat disimpulkan bahwa jenis segmentasi A pada kostumer memiliki jumlah data yang paling banyak. Kemudian diikuti dengan jenis segmentasi D pada urutan kedua, dan selanjutnya pada urutan ketiga ada jenis segmentasi B. Lalu untuk yang paling sedikit jumlah datanya, ada pada jenis segmentasi C

```python
segmentation_data = df["Segmentation"].value_counts()
x = segmentation_data.index
y = segmentation_data.values

plt.bar(x, y, color=['cyan','yellow','pink','blue'], label = ['A', 'D', 'B', 'C'])
plt.title('Jumlah Data Segmentasi Kostumer')
plt.xlabel('Jenis Segementasi')
plt.ylabel('Jumlah Data')
plt.legend()
plt.show()
```

## Visualisasi Frekuensi Kemunculan Data Pada Attribut Age Menggunakan Histogram
---
Dari visualisasi data pada histogram ini, dapat disimpulkan bahwa pada sekitar rentang umur 40-an memiliki jumlah frekuensi kemunculan data paling banyak, yang mana lebih dari 350 data. Sedangkan untuk yang paling sedikit frekuensi kemunculannya ada pada sekitar rentang umur 90-an, yang mana jumlah nya kurang dari 50 data

```python
sns.histplot(df['Age'], color = "blue", binwidth = 5, edgecolor = "white")
plt.title('Histogram Umur Kostumer')
plt.show()
```

## Visualisasi Distribusi Data Pada Attribute Spending_Score Menggunakan Pie Chart
---
Berdasarkan visualisasi data dalam pie chart di bawah ini, dapat disimpulkan bahwa data skor **High** memiliki jumlah distribusi data paling sedikit, yaitu sebesar 15% dari jumlah keseluruhan. Sementara itu, pada data skor **Average** memiliki jumlah distribusi data sebesar 24%. Dan untuk data distribusi terbanyak ada pada data skor **Low**, yang mana pada hal ini memiliki distribusi data sebesar 62%

```python
spending_score_data = df['Spending_Score'].value_counts()
plt.pie(spending_score_data, labels = spending_score_data.index, shadow = True, explode = [0, 0, 0.2], startangle = 90, autopct ='%.0f%%')
plt.title('Distribusi Data Skor Pengeluaran Kostumer')
plt.show()
```

## Visualisasi Korelasi Antar Data Pada Data Klasifikasi Segmentasi Kostumer Menggunakan Heatmap
---
Dari visualisasi data dalam heatmap yang ada pada gambar di bawah ini, dapat disimpulkan bahwa hampir setiap data memiliki korelasi yang cenderung negatif kepada data yang lainnya. Akan tetapi, dari visualisasi data ini bisa disimpulkan juga bahwa attribut **ID** memiliki pengaruh yang lumayan kuat kepada attribut **Family_Size**. Adapun attribut **Age** memiliki pengaruh yang paling lemah terhadap attribut **Family_Size**

```python
data_corr = df.corr()
sns.heatmap(data_corr, cmap="coolwarm", annot = True, center = 0)
plt.show()
```

## Visualisasi Persebaran Data Kostumer Segmentasi D Berdasarkan Umur dan Pengalaman Kerja Menggunakan Scatter Plot
---
Pada visualisasi data di bawah ini, dapat disimpulkan bahwa persebaran data kostumer segmentasi D dengan pengalaman kerja terbanyak ada pada rentang umur mulai dari sekitar 30-an hingga 40-an

```python
D = df.loc[df['Segmentation'] == 'D']
colScatter_x = 'Age'
colScatter_y = 'Work_Experience'
plt.scatter(x = D[colScatter_x], y = D[colScatter_y], color = 'blue', label='D')
plt.title('Persebaran Data Kostumer Segmentasi D Berdasarkan Umur dan Pengalaman Kerja')
plt.xlabel('Umur')
plt.ylabel('Pengalaman Kerja')
plt.legend()
plt.show()
```
